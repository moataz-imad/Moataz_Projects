```python
xo=0
xf=500
def calc(func='x^3-8*x-48',xo=xo,xf=xf,solved=True,acc=0.001):
    # return('hello')
    print(f'calc, {acc}')
    func0=func.replace('x','xo')
    func0=func0.replace('exop','exp')
    try:
        print(func0)
        sol0=eval(func0)
    except:
        print('error 1')
        return 0,False
    print(f'calc from {xo} to {xf}')
    
    funcf=func.replace('x','xf')
    funcf=funcf.replace('exfp','exp')
    solf=eval(funcf)
    if abs(solf)<acc:

        return xf,True
    
    if sol0*solf<=0:
        # root in between
        print('sol in between')
        xm=(xf+xo)/2
        funcm=func.replace('x','xm')
        funcm=funcm.replace('exmp','exp')
        try:
            solm=eval(funcm)
        except:
            print('error 2')
            return 0,False
        try:
            res=sol0*solm
            if abs(res)>100_000_000:
                print('error 3 approaching infinity')
                return 0,False
        except:
            print('error 4')
            return 0,False
            
        if res<=0:
            # root in between
            print('root in between')
            xf,solved=calc(func,xo=xo,xf=xm,solved=solved,acc=acc)
        else:
            xf,solved=calc(func,xo=xm,xf=xf,solved=solved,acc=acc)
    else:

        return 0,False
    return xf,True
# from calc import calc
func="1/x"
calc(func=func,xo=-0.06122,xf=0.1836)
```
