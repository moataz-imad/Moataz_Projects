```python
import requests
from bs4 import BeautifulSoup as bs
import time
import smtplib
import re
from enum import Enum
country='Jordan'
currency='USD'
url=f'https://www.numbeo.com/gas-prices/country_result.jsp?country={country}&displayCurrency={currency}'
headers={"User-Agent": 'Mozilla/5.0 (X11; Linux x86_64)', 'Cache-Control': 'no-cache', "Pragma": "no-cache"}
url_listofC='https://www.numbeo.com/gas-prices/'
data=requests.get(url=url,headers=headers)
data2=requests.get(url=url_listofC,headers=headers)
soup=bs(data.content)
souplist=bs(data2.content)
```

```python
value=re.findall(r'\d*.\d\d',soup.find_all(class_='first_currency')[0].text)[0]
```

```python
item=soup.find(class_='data_wide_table new_bar_table').find_all('tr')[1].find('td').text
```

```python
rows=[]
countriesList=souplist.find(class_='related_links').find_all('a')
for country in countriesList:
    r={}
    print(country.text)
    re.findall(r'.*',country.get('href'))
    s=country.get('href')
    cLabel=country.text
    cName=re.findall(r'try=(.*)',s)[0]
    r['Name']=country.text
    
    url0=f'https://www.numbeo.com/gas-prices/country_result.jsp?country={cName}&displayCurrency={currency}'
    headers={"User-Agent": 'Mozilla/5.0 (X11; Linux x86_64)', 'Cache-Control': 'no-cache', "Pragma": "no-cache"}
    data=requests.get(url=url0,headers=headers)
    soup=bs(data.content)
    try:
        r['cost']=float(re.findall(r'(\d*[\,\.\d]+\d)',soup.find_all(class_='first_currency')[0].text)[0])
    except:
        r['cost']='???'
    url02=f'https://www.numbeo.com/cost-of-living/country_result.jsp?country={cName}&displayCurrency=USD'
    data2=requests.get(url=url02,headers=headers)
    soup2=bs(data2.content)
    try:
        r['salary']=float(re.findall(r'(\d*[\,\.\d]+\d)',soup2.find(class_='data_wide_table new_bar_table').find_all('tr')[-2].find_all('td')[1].text)[0].replace(',',''))
    except:
        r['salary']='???'
    rows.append(r)
    
    # if len(rows)>15:
    #     break
```
