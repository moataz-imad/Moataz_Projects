```python
import os
import requests

def download_image(base_url, manga_name, chapter_start, chapter_end, page_start, page_end):
    try:
        # Create directory if it doesn't exist
        if not os.path.exists(manga_name):
            os.makedirs(manga_name)

        for chapter_num in range(chapter_start, chapter_end + 1):
            for page_num in range(page_start, page_end + 1):
                image_name = f"{chapter_num:04}-{page_num:03}.png"
                image_url = f"{base_url}/{chapter_num:04}-{page_num:03}.png"
                response = requests.get(image_url)
                if response.status_code == 200:
                    save_path = os.path.join(manga_name, image_name)
                    with open(save_path, 'wb') as f:
                        f.write(response.content)
                    print(f"Image {image_name} downloaded successfully.")
                else:
                    if page_num>20:
                        continue
                    print(f"Image {image_name} not found.")
    except Exception as e:
        pass


base_url = "manga_website"
base_url = "manga_website2"
manga_name = "Jujutsu-Kaisen" 
chapter_start = 190
chapter_end = 250
page_start = 1
page_end = 30

download_image(base_url, manga_name, chapter_start, chapter_end, page_start, page_end)
```

```python
import os
import img2pdf

def get_all_png_files(directory):
    png_files = {}
    for root, _, files in os.walk(directory):
        for file in files:
            if file.endswith(".png"):
                chapter = int(file.split("-")[0])  # Extract chapter number from filename
                group = (chapter - 1) // 10  # Calculate the group number
                group_start = group * 10 + 1  # Start chapter of the group
                group_end = min((group + 1) * 10, group_start + 9)  # End chapter of the group
                group_key = f"{group_start:04d}-{group_end:04d}"
                if group_key not in png_files:
                    png_files[group_key] = []
                png_files[group_key].append(os.path.join(root, file))
    return png_files

def convert_images_to_pdf(image_paths, output_pdf):
    with open(output_pdf, "wb") as f:
        f.write(img2pdf.convert([open(image, 'rb') for image in image_paths]))

def create_pdfs_by_group(image_paths, output_directory):
    for group, images in image_paths.items():
        pdf_name = f"{group}.pdf"
        output_pdf = os.path.join(output_directory, pdf_name)
        convert_images_to_pdf(images, output_pdf)

if __name__ == "__main__":
    directory = "Jujutsu-Kaisen" 
    output_directory = "Grouped_PDFs"
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)
    
    image_paths = get_all_png_files(directory)
    create_pdfs_by_group(image_paths, output_directory)
```
