```python
from selenium import webdriver
path='../ChromeDriver/chromedriver.exe'
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import random
import pandas as pd
```

```python
try:
    wd.quit()
except:
    pass
wd=webdriver.Chrome(path)
wd.get('airlinewebsite.com') # hidden for legal reasons
```

```python
df=pd.DataFrame()
def scroll_to_end(wd=wd):
        wd.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(1) 
def scroll_to_top(wd=wd):
        wd.execute_script("window.scrollTo(0, 0);")
        time.sleep(1) 

def get_dur(duration='04h36m'):
    h,m=duration.replace('m','').split('h')
    h=float(h)+float(m)/60
    return round(h,2)
def get_info(flights,wd=wd,df=df):
    fares_labels=wd.find_elements(By.CLASS_NAME,'farefamily-header-container')[0].text.split('\n') # costs labels ['Economy Flex', 'Business Plus']
    for fid,f in enumerate(flights):
        data={}
        segments=f.find_elements(By.CLASS_NAME,'timeline-segment')
        if len(segments)==0:
            f.find_element(By.CLASS_NAME,'availability-show-timeline').click()
        data['Title'],data['Date']=f.find_element(By.CLASS_NAME,'timeline-header').text.split(',') # Amman - San Francisco, Wednesday 19 July
        data['From Airport']=f.find_element(By.CLASS_NAME,'airportname-from').text
        data['To Airport']=f.find_element(By.CLASS_NAME,'airportname-to').text
        data['Stops']=int(f.find_element(By.CLASS_NAME,'stop').text.replace('stop',''))
        dur=f.find_element(By.CLASS_NAME,'duration').text.replace('Duration','')
        data['Total Duration [Hours]'] = get_dur(dur)
        costs=f.find_elements(By.CLASS_NAME,'cell-reco')
        for ci,c in enumerate(costs):
            price=c.text.replace('JOD','').replace('Lowest Fares','').replace('\n','').replace(',','')
            if 'Sold Out' in price:
                price = float('nan')
            else:
                price=float(price)
            label=f'{fares_labels[ci]} [JOD]'
            data[label]=price
        
        segments=f.find_elements(By.CLASS_NAME,'timeline-segment')
        for si,s in enumerate(segments):
            print('##########')
            try:
                dur =s.find_elements(By.CLASS_NAME,'timeline-info-stopduration')[1].text # for the stopping 
                data[f'{si}_stop-duration [Hours]']=get_dur(dur)
            except:
                data[f'{si}_takeoff-location']=s.find_elements(By.CLASS_NAME,'timeline-takeoff-locationcode')[0].get_attribute('title')
                data[f'{si}_takeoff-loc']=s.find_elements(By.CLASS_NAME,'timeline-takeoff-locationcode')[0].text
                data[f'{si}_landing-location']=s.find_elements(By.CLASS_NAME,'timeline-landing-locationcode')[0].get_attribute('title')
                data[f'{si}_landing-loc']=s.find_elements(By.CLASS_NAME,'timeline-landing-locationcode')[0].text
                data[f'{si}_departure-time']=s.find_elements(By.CLASS_NAME,'timeline-time-departure')[0].text
                data[f'{si}_arrival-time']=s.find_elements(By.CLASS_NAME,'timeline-time-arrival')[0].text
                dur=s.find_elements(By.CLASS_NAME,'timeline-flight-duration-textdate')[0].text
                data[f'{si}_flight-duration [Hours]']=get_dur(dur)
        scroll_to_end()
        df0=pd.DataFrame(data,index=range(1))
        df=pd.concat([df,df0],ignore_index=True)
        print('__________________________')
    scroll_to_top()
    return df
```
